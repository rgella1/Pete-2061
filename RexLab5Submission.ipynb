2. **Using the example in "lab5_Probability.ipynb", write a Python code to compute the probabililities of all possible total values that can be obtained from rolling a fair die three times**

```python
import itertools
import pandas as pd
import numpy as np
x = list(range(1,7))
y = list(itertools.product(x,repeat=int(input("Number of Dice = "))))
z = np.array(y)

print(f"Total number of outcomes = {len(y)}")
sumOutcomes = np.sum(z, axis=1)
d = pd.DataFrame({"outcomes":y,"sumOutcomes":sumOutcomes})
d

by_sumOutcomes = d.groupby("sumOutcomes")
by_sumOutcomes.describe()
outcomeDF = by_sumOutcomes.count()
outcomes = outcomeDF.values[:,0]
probability = (outcomes / np.sum(outcomes))*100
f = pd.DataFrame({"Sum of Three Dice":d["sumOutcomes"].unique(), "Outcomes":outcomes, "% Probability":probability})
f
```
